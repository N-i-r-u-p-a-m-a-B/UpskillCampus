```python
from scipy import stats
```

```python
# assign data
x = data[' 3-2003']
y = data[' 3-2004']


slope, intercept, r, p, std_err = stats.linregress(x, y)


# function to return slope
def myfunc(x):
    return slope * x + intercept


mymodel = list(map(myfunc, x))

# scatter
plt.scatter(x, y)

# plotting the data
plt.plot(x, mymodel)

# display the figure
plt.show()
```
